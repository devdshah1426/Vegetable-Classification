# **Bell** **Pepper**

Top-5 accuracy: 82.92%

Top-1 accuracy: 52.00%

```python
# DAMAGED 
from keras.applications.inception_v3 import InceptionV3
import keras.utils as image
from keras.applications.inception_v3 import preprocess_input, decode_predictions
import numpy as np
import tqdm

top1_acc = 0
top5_acc = 0
data_dir = r"/content/gdrive/MyDrive/Colab Notebooks/KU/Dataset/Damaged Bell Pepper"

# Load the InceptionV3 model
model = InceptionV3(weights='imagenet')

# Iterate over all images in the data directory
for i in tqdm.tqdm(range(1, 32)):
    img_path = data_dir + '/Damaged_' + str(i) + '.jpg'

    # Load the image and preprocess it for the model
    img = image.load_img(img_path, target_size=(299, 299))

    # Convert the image to a numpy array
    x = image.img_to_array(img)

    # Add an extra dimension to the array (as required by the model)
    x = np.expand_dims(x, axis=0)

    # Preprocess the image by scaling the pixel values
    x = preprocess_input(x)

    # Run the image through the model
    predictions = model.predict(x)
    decoded_predictions = decode_predictions(predictions, top=5)[0]

    top1_pred = decode_predictions(predictions, top=1)[0][0][1]
    if top1_pred == 'bell_pepper':
        top1_acc += 1
    for pred in decoded_predictions:
        if pred[1] == 'bell_pepper':
            top5_acc += 1
            break

# calculate the top-5 accuracy
top5_acc = top5_acc / (32 * 1.0)
print("Top-5 accuracy: {:.2f}%".format(top5_acc * 100))

# calculate the top-1 accuracy
top1_acc = top1_acc / (32 * 1.0)
print("Top-1 accuracy: {:.2f}%".format(top1_acc * 100))
```
