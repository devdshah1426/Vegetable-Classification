# **Bell Pepper**

Top-5 accuracy: 95.24%

Top-1 accuracy: 72.47%


This script uses the Xception model, which is a deep convolutional neural network model trained on the ImageNet dataset, to classify images of bell peppers. It loads the pre-trained model and then iterates over all images in a specified directory, preprocesses the images, runs them through the model, and decodes the top 5 predictions for each image. It then checks if the first or any of the top 5 predictions match the label 'bell_pepper' and keeps track of the number of times this occurs. Finally, it calculates and prints the top-1 and top-5 accuracy of the model in classifying the images as bell peppers.

```python
# DAMAGED 
import keras.utils as image
from keras.applications.xception import Xception
#from keras.preprocessing import image
from keras.applications.xception import preprocess_input, decode_predictions
import numpy as np
import tqdm

top1_acc = 0
top5_acc = 0
data_dir = r"/content/gdrive/MyDrive/Colab Notebooks/KU/Dataset/Damaged Bell Pepper"

# Load the Xception model
model = Xception(weights='imagenet')

# Iterate over all images in the data directory
for i in tqdm.tqdm(range(1, 32)):
    img_path = data_dir + '/Damaged_' + str(i) + '.jpg'

    # Load the image and preprocess it for the model
    img = image.load_img(img_path, target_size=(299, 299))

    # Convert the image to a numpy array
    x = image.img_to_array(img)

    # Add an extra dimension to the array (as required by the model)
    x = np.expand_dims(x, axis=0)

    # Preprocess the image by scaling the pixel values
    x = preprocess_input(x)

    # Run the image through the model
    predictions = model.predict(x)
    decoded_predictions = decode_predictions(predictions, top=5)[0]

    top1_pred = decode_predictions(predictions, top=1)[0][0][1]
    if top1_pred == 'bell_pepper':
        top1_acc += 1
    for pred in decoded_predictions:
        if pred[1] == 'bell_pepper':
            top5_acc += 1
            break

# calculate the top-5 accuracy
top5_acc = top5_acc / (32 * 1.0)
print()
print("Top-5 accuracy: {:.2f}%".format(top5_acc * 100))

# calculate the top-1 accuracy
top1_acc = top1_acc / (32 * 1.0)
print("Top-1 accuracy: {:.2f}%".format(top1_acc * 100))
```
