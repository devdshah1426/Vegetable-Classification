# **Bell Pepper**

Top-5 accuracy: 0.00%

Top-1 accuracy: 0.00%

```python
import keras.utils as image
from keras.applications.vgg16 import VGG16
from keras.applications.xception import preprocess_input, decode_predictions
import numpy as np
import tqdm

top1_acc = 0
top5_acc = 0
data_dir = r"/content/gdrive/MyDrive/Colab Notebooks/KU/Dataset/bellpeper"

# Load the VGG16 model
model = VGG16(weights='imagenet')

# Iterate over all images in the data directory
for i in tqdm.tqdm(range(1, 1177)):
    img_path = data_dir + '/bellpeper_' + str(i) + '.jpg'

    # Load the image and preprocess it for the model
    img = image.load_img(img_path, target_size=(224, 224))

    # Convert the image to a numpy array
    x = image.img_to_array(img)

    # Add an extra dimension to the array (as required by the model)
    x = np.expand_dims(x, axis=0)

    # Preprocess the image by scaling the pixel values
    x = preprocess_input(x)

    # Run the image through the model
    predictions = model.predict(x)
    decoded_predictions = decode_predictions(predictions, top=5)[0]

    # Convert the predictions to class labels
    label_predictions = decode_predictions(predictions, top=5)
    print(label_predictions)

    top1_pred = decode_predictions(predictions, top=1)[0][0][1]
    if top1_pred == 'bell_pepper':
        top1_acc += 1
    for pred in decoded_predictions:
        if pred[1] == 'bell_pepper':
            top5_acc += 1
            break

# calculate the top-5 accuracy
top5_acc = top5_acc / (1177 * 1.0)
print("Top-5 accuracy: {:.2f}%".format(top5_acc * 100))

# calculate the top-1 accuracy
top1_acc = top1_acc / (1177 * 1.0)
print("Top-1 accuracy: {:.2f}%".format(top1_acc * 100))
```
